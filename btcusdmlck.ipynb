```python
import requests as r
import pandas as pd 
import numpy as np
import json 
#import pandasql as ps
import matplotlib.pyplot as plt
```

```python
response  = r.get("https://api.coingecko.com/api/v3/coins/bitcoin/ohlc?vs_currency=usd&days=30").json()
print (response)
```

```python
res_df = pd.DataFrame(response)
print (res_df)
```

```python
X = res_df.iloc[:,[0,1,2,3]].values
Y = res_df.iloc[:,[4]].values
print (Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
Y_pred = lr.predict(X_test)
print (Y_pred)
```

```python
Yloc = res_df.iloc[:,[4]]
yluc = Yloc.iloc[:61].values
#print (ylu
avg = (Y_pred / yluc) 
print (avg)
```
